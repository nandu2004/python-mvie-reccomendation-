```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
# ast.literal_eval(text): This part of the function takes a string and evaluates it as a Python literal. This means it parses the string and returns a Python data structure. For example, if the input string is '[1, 2, 3]', the ast.literal_eval() function will return a list containing the integers 1, 2, and 3 stackoverflow.com.
# for i in ast.literal_eval(text): This part of the function iterates over each item in the list returned by ast.literal_eval(text).
# L.append(i['name']): For each item in the list, the function tries to append the value of the 'name' key to a new list L.
# return L: The function finally returns the new list L.
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
    # It initializes an empty list L and a counter variable counter.
# It then evaluates the input string text into a Python literal (probably a list of dictionaries), and iterates over each item in the list.
# For each item, if the counter is less than 3, it appends the 'name' value from the dictionary to the list L and increments the counter.
# Finally, it returns the list L.
```

```python
 movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['keywords'] + movies['cast'] + movies['crew'] + movies['genres']
```

```python
new = movies.drop(columns=['overview','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
pa= PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(pa.stem(i))

  return " ".join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
new['genres'] = new['genres'].apply(lambda x: " ".join(x))
```

```python
new.head()
```

```python
vector2 = cv.fit_transform(new['genres']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity2 = cosine_similarity(vector2)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])[1:6]
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
def recommendgenre(genre):
    index = new[new['genres'] == genre].index[0]
    distances2 = sorted(list(enumerate(similarity2[index])),reverse=True,key = lambda x: x[1])[1:20]
    for i in distances2[1:20]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommendgenre('Action')
```
